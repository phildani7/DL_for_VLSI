```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler # (0-1)
```

```python
#Generate data
X = np.random.randint(1,10, 1000)
```

```python
X = X.reshape(-1,1)
```

```python
y =np.array([int(x % 2 == 0) for x in X])
```

```python
#Split data into train & test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
#Scale data 0 to 1
scaler = MinMaxScaler(feature_range=(0,1))
X_train=scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(16, input_dim=1, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))

# Ac(wx+b)  | | |
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])

```

```python
model.fit(X_train, y_train, epochs=5000, batch_size=100)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
```

```python
new_data = np.random.randint(1,10, 1000)
new_data = new_data.reshape(-1,1)

#Scale data 0 to 1
scaler = MinMaxScaler(feature_range=(0,1))
new_data_train=scaler.fit_transform(new_data)

predictions = model.predict(new_data_train)
```
