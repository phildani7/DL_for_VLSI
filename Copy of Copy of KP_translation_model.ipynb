```python
import embeddings
import numpy as np
from cupy_utils import *
```

```python
src_embeddings = '/content/drive/MyDrive/KP-Translation_Model/SRC_Mapped_Hi-KP.emb'
trg_embeddings = '/content/drive/MyDrive/KP-Translation_Model/TRG_Mapped_Hi-KP.emb'
```

```python
srcfile = open(src_embeddings, encoding='utf-8', errors='surrogateescape')
trgfile = open(trg_embeddings, encoding='utf-8', errors='surrogateescape')
src_words, x = embeddings.read(srcfile, dtype='float32')
trg_words, z = embeddings.read(trgfile, dtype='float32')
```

```python
embeddings.length_normalize(x)
embeddings.length_normalize(z)
```

```python
xp = get_cupy()
x = cupy.asarray(x)
z = cupy.asarray(z)
```

```python
sr_word2emb = {}
for word, vector in zip(src_words, x):
  sr_word2emb[word] = vector

tr_word2emb = {}
for word, vector in zip(trg_words, z):
  tr_word2emb[word] = vector
```

```python
testWord = 'मौका' #आप, 

similarities = sr_word2emb[testWord].dot(z.T)

# 10 words retrieval using Nearest Neighbor
K = 10
nn = similarities.argsort().tolist()[:K]
for kp in nn :
  print(trg_words[kp])
```

```python
# 10 words retrieval using K-means Nearest Neighbor
knn = (-similarities).argsort().tolist()[:K]
for kn in knn :
  print(trg_words[kn])
```
