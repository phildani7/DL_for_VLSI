```python
#Initial values

weight = 1.0
bias = 0.0
input = 2
#target = 10
target = 20
#learning_rate = 0.1
learning_rate = 0.01
```

```python
#Forward Propagation
output = weight * input + bias
#2 = 1 * 2 + 0
```

```python
output
```

```python
#Calculate error (MSE)
error = (target - output) ** 2
```

```python
error
```

```python
for i in range(200):
  output = weight * input + bias # Forward Prop
  error = (target - output) ** 2 # Calculate Error
  dError_dOutput = -2 * (target - output)
  dError_dWeight = dError_dOutput * input #weights - > derivative
  dError_dBias = dError_dOutput #bias -> derivative
  weight = weight - learning_rate * dError_dWeight #Applying new weights
  bias = bias - learning_rate * dError_dBias #Applying new bias
  print(f"After Epoch {i+1}: Output = {output}, Error = {error}, Weight = {weight}, Bias = {bias}")
```

```python
#converge after 200 epochs; .999 accuracy
```
