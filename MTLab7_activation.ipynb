```python
import numpy as np
import matplotlib.pyplot as plt

#SIGMOID
def sigmoid(x):
  return 1/ (1+np.exp(-x))
```

```python
x = np.linspace(-10, 10, 400)
plt.figure(figsize=(14,8))
plt.subplot(2,2,1)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid Activation')
plt.grid()
```

```python
#x -> centered around zero
#y -> 0-1
#steep changes around x =0
#Activation -> always +ve
#
#Vanishing gradient problem
#NO: Not for hidden layers
#NO: Multiple output classification
#YES: Output - Binary classification = Inividual probability (0/1)


#output=Sigmoid(wx+b)
```

```python
#2 TANH

def tanh(x):
  return np.tanh(x)

plt.subplot(2, 2, 2)
plt.plot(x,tanh(x))
plt.title('Tanh Activation')
plt.grid()
```

```python
#x & y -> centered around zero
#y -> -1 to +1
#steeper than sigmoid: changes around x =0
#Activation -> -ve and +ve
#
#Vanishing gradient problem
#zig-zag learning
#NO: Not for hidden layers
#NO: Binary classification, Multiple output classification
#YES: Regression, bipolar data , RNN, GAN, Autoencoders
```

```python
#RELU
def relu(x):
  return np.maximum(0,x)

plt.subplot(2, 2, 3)
plt.plot(x,relu(x))
plt.title('ReLU Activation')
plt.grid()
```

```python
#Very good - All layers
#NO Vanishing gradient problem
#Derivative = 0 or 1
#Very fast, simple math.

#Activation ->  +ve
#Cons: Leads to sparsity (eliminates -ve values) (100 weights -> 30 are zero)
# Training could get stuck -> "dying ReLU"
# Eery growing activation - no limit

#zig-zag learning
#NO: Not for hidden layers
#NO: Binary classification, Multiple output classification
#YES: Regression, bipolar data , RNN, GAN, Autoencoders
```
