```python
def maximum(value1, value2, value3):
  """Return the maximum of three values. """
  max_value = value1
  if value2 > max_value:
    max_value = value2
  if value3 > max_value:
    max_value = value3
    return max_value
```

```python
maximum(5.8, 8.7, 6.9)
```
