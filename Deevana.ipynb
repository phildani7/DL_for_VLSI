```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx


# Define groups
groups = {
    'context': ['Q1', 'Q2', 'Q3', 'Q4'],
    'planning': ['Q5', 'Q6', 'Q7', 'Q8'],
    'input': ['Q9', 'Q10', 'Q11', 'Q12', 'Q13', 'Q14', 'Q15', 'Q16'],
    'process': ['Q17', 'Q18', 'Q19', 'Q20', 'Q21'],
    'output': ['Q22', 'Q23', 'Q24', 'Q25'],
    'outcome': ['Q26', 'Q27', 'Q28', 'Q29']
}

# Group visualization
plt.figure(figsize=(14, 10))

for group_name, qs in groups.items():
    avg_scores = df.loc[qs].mean(axis=1)
    sns.barplot(x=avg_scores.index, y=avg_scores.values, label=group_name)

plt.title('Average Scores Across Different Groups')
plt.xlabel('Questions')
plt.ylabel('Average Score')
plt.legend(loc='upper right')
plt.show()

# Define the relationships
relationships = {
    'Q2': ['Q1', 'Q3', 'Q4', 'Q6', 'Q15'],
    'Q3': ['Q4'],
    'Q6': ['Q5', 'Q2', 'Q7', 'Q8', 'Q9', 'Q10', 'Q15', 'Q16', 'Q34', 'Q35'],
    'Q15': ['Q5', 'Q11'],
    'Q11': ['Q12', 'Q13', 'Q14'],
    'Q19': ['Q17', 'Q18', 'Q20', 'Q30', 'Q31', 'Q32'],
    'Q21': ['Q22', 'Q27'],
    'Q16': ['Q2', 'Q6', 'Q17', 'Q27', 'Q28', 'Q29'],
    'Q22': ['Q33'],
    'Q34': ['Q2', 'Q6', 'Q15', 'Q16'],
    'Q36': ['Q2', 'Q16', 'Q15'],
    'Q38': ['Q23', 'Q39', 'Q40']
}

G = nx.DiGraph()

for main_q, related_qs in relationships.items():
    for q in related_qs:
        G.add_edge(main_q, q)

pos = nx.spring_layout(G)
plt.figure(figsize=(14, 10))
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, width=1.5, font_size=15, arrowsize=20)
plt.title('Relationships Among Questions')
plt.show()
```

Certainly! The relationship dictionary describes directed dependencies between questions. By examining the structure of these dependencies, you can extract several types of inferences:

1. **Centrality Measures**: By checking which questions are most often referenced by other questions, you can determine which questions are central or pivotal to the set. This is often referred to as degree centrality in network analysis.

2. **Dependency Chains**: You can identify sequences of questions that depend on each other. For instance, if Q2 depends on Q1, and Q3 depends on Q2, you have a chain from Q1 to Q2 to Q3.

3. **Cycles**: These are questions that depend on each other in a circular way. If, for example, Q2 depends on Q1, and Q1 also depends on Q2, then there's a cycle. Cycles might indicate areas where clarification or further breakdown is required.

4. **Isolated Questions**: Questions that neither depend on others nor are they depended upon can be identified. These might be stand-alone topics or areas which aren't yet fully integrated into the structure.

5. **Clusters or Communities**: Using community detection algorithms, it's possible to detect clusters of closely related questions. This can help in grouping topics or themes.

6. **Bottlenecks**: These are questions that, if not addressed, block a large number of other questions from being addressed. Identifying bottlenecks can help prioritize which questions to address first.



```python
import networkx as nx

G = nx.DiGraph()

for main_q, related_qs in relationships.items():
    for q in related_qs:
        G.add_edge(main_q, q)

# Compute degree centrality:
degree_centrality = nx.degree_centrality(G)

# Sort by centrality
sorted_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)

print("Questions ranked by their centrality:")
for q, centrality in sorted_degree:
    print(f"Question {q} has centrality of {centrality:.2f}")
```

```python
# Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
sorted_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)

print("Questions ranked by their betweenness centrality:")
for q, centrality in sorted_betweenness:
    print(f"Question {q} has betweenness centrality of {centrality:.2f}")
```

```python
all_paths = []
for node in G:
    for path in nx.all_simple_paths(G, source=node, target=sorted_degree[0][0]):  # target is most central node
        all_paths.append(path)

# Sort paths by length
all_paths = sorted(all_paths, key=len, reverse=True)
print("Longest dependency chains:")
for path in all_paths[:5]:  # Print top 5 longest paths
    print(" -> ".join(path))
```

```python
cycles = list(nx.simple_cycles(G))
print("Detected cycles:")
for cycle in cycles:
    print(" -> ".join(cycle))
```

```python
from networkx.algorithms import community
communities = community.greedy_modularity_communities(G.to_undirected())
print("Detected communities:")
for i, com in enumerate(communities, 1):
    print(f"Community {i}: {', '.join(com)}")
```
