```python
# Define the initial values
weight = 1.0
bias = 0.0
input = 2
target = 10
learning_rate = 0.1

# Forward Propagation
output = weight * input + bias
print(f"Initial Output: {output}")

# Calculate Error
error = (target - output) ** 2
print(f"Initial Error: {error}")

# Start Training (let's say we do this for 20 epochs)
for i in range(20):
    # Forward Propagation
    output = weight * input + bias

    # Calculate Error
    error = (target - output) ** 2

    # Backpropagation
    dError_dOutput = -2 * (target - output)
    dError_dWeight = dError_dOutput * input
    dError_dBias = dError_dOutput

    # Update weights and bias
    weight = weight - learning_rate * dError_dWeight
    bias = bias - learning_rate * dError_dBias

    print(f"After Epoch {i+1}: Output = {output}, Error = {error}, Weight = {weight}, Bias = {bias}")
```
