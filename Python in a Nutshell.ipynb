```python
from scipy.signal import butter, lfilter

def lowpass_filter(data, cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    filtered_data = lfilter(b, a, data)
    return filtered_data

# Sample data
data = np.sin(2 * np.pi * 1.2 * np.linspace(0, 1, 100)) + np.random.randn(100) * 0.2

# Apply filter
filtered_data = lowpass_filter(data, cutoff=0.5, fs=10)

# Plot
plt.plot(data, label='Noisy Signal')
plt.plot(filtered_data, label='Filtered Signal')
plt.legend()
plt.show()
```

```python
from scipy.signal import spectrogram

# Generate a signal
fs = 10e3
N = 1e5
amp = 2 * np.sqrt(2)
freq = 1234.0
noise_power = 0.001 * fs / 2
time = np.arange(N) / fs
signal = amp * np.sin(2 * np.pi * freq * time)
signal += np.random.normal(scale=np.sqrt(noise_power), size=time.shape)

# Compute and plot the spectrogram
frequencies, times, Sxx = spectrogram(signal, fs)
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.title("Spectrogram")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

g = 9.81  # Gravity constant
v0 = 20   # Initial velocity (m/s)
angle = 45  # Launch angle in degrees

# Time of flight calculation
t_flight = (2 * v0 * np.sin(np.radians(angle))) / g

# Time intervals
t = np.linspace(0, t_flight, num=100)

# X and Y coordinates
x = v0 * t * np.cos(np.radians(angle))
y = v0 * t * np.sin(np.radians(angle)) - 0.5 * g * t**2

plt.plot(x, y)
plt.title("Projectile Motion")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.show()
```

```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Logistic growth model dP/dt = rP(1 - P/K)
def logistic_growth(t, P, r, K):
    return r * P * (1 - P/K)

# Parameters
r = 0.1  # Growth rate
K = 1000 # Carrying capacity
P0 = 10  # Initial population

sol = solve_ivp(logistic_growth, [0, 100], [P0], args=(r, K), dense_output=True)

t = np.linspace(0, 100, 100)
P = sol.sol(t)

plt.plot(t, P.T)
plt.title("Logistic Growth Model")
plt.xlabel("Time")
plt.ylabel("Population")
plt.show()
```

```python
from astropy.constants import G
import astropy.units as u
from astropy import time

# Assume some typical values for a planet
mass_sun = 1.989 * 10**30 * u.kg
mass_planet = 5.972 * 10**24 * u.kg
distance = 149.6 * 10**6 * u.km

# Orbital period calculation using Kepler's third law
T = np.sqrt(4 * np.pi**2 * distance**3 / (G * (mass_sun + mass_planet)))

print("Orbital Period:", T.to(u.year))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Simulate noisy sensor data
t = np.linspace(0, 1, 100)
sensor_data = np.sin(2 * np.pi * 10 * t) + np.random.normal(0, 0.5, 100)

# Butterworth filter
b, a = butter(3, 0.05)
filtered_data = lfilter(b, a, sensor_data)

plt.plot(t, sensor_data, label='Noisy')
plt.plot(t, filtered_data, label='Filtered')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def simulate_flight(velocity, angle, time):
    g = 9.81  # Gravity constant
    angle_rad = np.radians(angle)

    # Equations of motion
    x = velocity * time * np.cos(angle_rad)
    y = velocity * time * np.sin(angle_rad) - 0.5 * g * time**2
    return x, y

# Sample parameters
velocity = 300  # m/s
angle = 45  # degrees
time = np.linspace(0, 30, 300)  # Simulation time

x, y = simulate_flight(velocity, angle, time)

# Plot
plt.plot(x, y)
plt.title("Flight Trajectory")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.show()
```

```python
import numpy as np

def calculate_stress(force, area):
    return force / area

# Sample data
force = 1000  # Newton
area = 0.01  # Square meters

stress = calculate_stress(force, area)
print(f"Stress: {stress} Pa")
```

```python
def heat_transfer(coefficient, area, temp_difference):
    return coefficient * area * temp_difference

# Sample parameters
coefficient = 10  # Heat transfer coefficient (W/m^2K)
area = 2  # m^2
temp_difference = 40  # Temperature difference in Kelvin

heat = heat_transfer(coefficient, area, temp_difference)
print(f"Heat Transfer: {heat} Watts")
```

```python
def calculate_lift(coefficient, density, velocity, area):
    return 0.5 * coefficient * density * velocity**2 * area

# Sample parameters
coefficient = 1.2  # Lift coefficient
density = 1.225  # Air density at sea level (kg/m^3)
velocity = 250  # Velocity (m/s)
area = 20  # Wing area (m^2)

lift = calculate_lift(coefficient, density, velocity, area)
print(f"Lift Force: {lift} Newtons")
```
