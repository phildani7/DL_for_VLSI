```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
np.random.seed(0)
data = np.random.rand(1000, 1) * 100
categories = np.digitize(data, bins=[33.33, 66.66])
labels = to_categorical(categories)
```

```python
data[0]
```

```python
labels[0]
```

```python
split_index = int(0.8 * 1000)
X_train, X_test = data[:split_index], data[split_index:]
y_train, y_test = labels[:split_index], labels[split_index:]
```

```python
X_train[0]
```

```python
y_train
```

```python
model = Sequential([
    Dense(64, input_shape=(1,), activation='relu'),
    Dense(128, activation='relu'),
    Dense(3, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=300, batch_size=64, validation_data=(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```
