```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler #(0-1)
```

```python
np.random.seed(45)
X = np.random.randint(1,10,1000)
```

```python
X
```

```python
X = X.reshape(-1,1)
```

```python
X.shape
```

```python
X[0]
```

```python
y = np.array([int(x % 2 == 0) for x in X])
```

```python
y[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(16, input_dim=1, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='relu'))
```

```python
model.summary()
```

```python
model.compile(loss='mse',optimizer='adam',metrics=['acc'])
```

```python
model.fit(X_train, y_train, epochs=5000, batch_size=100)
```

```python
new_data = np.random.randint(1,10,1000)
```

```python
new_data= new_data.reshape(-1,1)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
new_data_train=scaler.fit_transform(new_data)
```

```python
predictions = model.predict(new_data_train)
```
