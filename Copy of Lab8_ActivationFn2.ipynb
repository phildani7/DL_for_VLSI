```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)  # only returns the first class probability

#YES: When RELU is not training and GPU resource available
#Addresses dying ReLU problem
#Saturation for negative region helps with the "exploding gradient" problem.
#Can produce negative outputs.
#NO:
#Computationally more expensive than ReLU because of the exponential function.

def elu(x, alpha=1.0):
    return np.where(x > 0.0, x, alpha * (np.exp(x) - 1))


#YES:
#Can handle negative values better than ReLU.
#Has been found to outperform ReLU in deeper models.
#NO: Computationally more intensive than ReLU.
def swish(x, beta=1.0):
    return x / (1 + np.exp(-beta * x))

def mish(x):
    return x * np.tanh(np.log1p(np.exp(x)))

def prelu(x, alpha=0.25):  # setting alpha as 0.25 for illustration
    return np.where(x > 0, x, alpha * x)

# Generate data
x = np.linspace(-10, 10, 400)

# Plotting
plt.figure(figsize=(18, 12))

plt.subplot(3, 2, 1)
plt.plot(x, softmax(x))
plt.title('Softmax Activation (For 1st class)')
plt.grid()

plt.subplot(3, 2, 2)
plt.plot(x, elu(x))
plt.title('ELU Activation')
plt.grid()

plt.subplot(3, 2, 3)
plt.plot(x, prelu(x))
plt.title('PReLU Activation')
plt.grid()

plt.subplot(3, 2, 4)
plt.plot(x, swish(x))
plt.title('Swish Activation')
plt.grid()

plt.subplot(3, 2, 5)
plt.plot(x, mish(x))
plt.title('Mish Activation')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softplus(x):
    return np.log(1 + np.exp(x))

def mish(x):
    return x * np.tanh(softplus(x))

def swish(x):
    return x * sigmoid(x)

x = np.linspace(-5, 5, 400)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, swish(x), label='Swish')
plt.title('Swish Activation')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, mish(x), label='Mish')
plt.title('Mish Activation')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
